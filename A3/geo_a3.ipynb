```python
# Population data for cities in Austria

aust_pop = pd.read_csv("dataset/at.csv")
aust_pop.head()
```

```python
# Plotting the Austrian population data using markers
from ipyleaflet import Marker
m = Map(center = ( 47.6682, 14.5501), zoom=7, basemap = basemaps.Esri.WorldStreetMap)
for city, lat, lon in aust_pop[["city", "lat", "lng"]].values:
    marker = Marker(location=(lat, lon), draggable=False, title=city, alt=city)
    m.add_layer(marker)
m  
```

```python
# importing classes for full screen, scale control
from ipyleaflet import Map, FullScreenControl, LayersControl, MeasureControl, ScaleControl

# add full screen control, default position is top left
m.add_control(FullScreenControl())

# add scale control
m.add_control(ScaleControl(position="bottomleft"))
m
```

```python
# Create clusters for markers to solve the problem of overcrowding

from ipyleaflet import MarkerCluster
m = Map(center = (47.6682, 14.5501), zoom=7, basemap = basemaps.OpenStreetMap.HOT)

markers = []
for city, lat, lon in aust_pop[["city", "lat", "lng"]].values:
    markers.append(Marker(location=(lat, lon), draggable=False, title=city, alt=city))
    
marker_cluster = MarkerCluster(markers=markers)
m.add_layer(marker_cluster)
m
```
